假設我們資料中類別的數量並不均衡，在評估準確率時可能會有所偏頗，試著切分出 y_test 中，0 類別與 1 類別的數量是一樣的 (亦即 y_test 的類別是均衡的)

```python
import numpy as np
X = np.arange(1000).reshape(200, 5)
y = np.zeros(200)
y[:40] = 1
```

```python
y
```

可以看見 y 類別中，有 160 個 類別 0，40 個 類別 1 ，請試著使用 train_test_split 函數，切分出 y_test 中能各有 10 筆類別 0 與 10 筆類別 1 。(HINT: 參考函數中的 test_size，可針對不同類別各自作切分後再合併)

```python
from sklearn.model_selection import train_test_split, KFold
import numpy as np
```

```python
y_1=[]
for i in y:
    if i==1:
        y_1.append(i)
y_1
```

```python
y
```

```python
y_0=[]
for i in y:
    if i==0:
        y_0.append(i)
y_0
```

```python
train_data_y1, test_data_y1, train_label_y1, test_label_y1 = train_test_split(X[:len(y_1)], y_1, test_size=10)
train_data_y0, test_data_y0, train_label_y0, test_label_y0 = train_test_split(X[:len(y_0)], y_0, test_size=10)
```

```python
y_test = np.concatenate([test_label_y1, test_label_y0])
```
